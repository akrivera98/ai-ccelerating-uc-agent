```python
from data_utils import get_input_data, input_data_summary, diff_instance, check_storage_status
import random
```

```python
data_dir = "../data/starting_kit/Train_Data"
```

```python
data_in, data_out = get_input_data(data_dir)
```

```python
for case in list(data_in.keys())[:5]:
    summary = input_data_summary(data_in[case])
    print(f"Case: {case}")
    print(f"Summary: {summary}")
    print("-" * 50)
```

```python
# Pick two random cases to compare
instance_names = list(data_in.keys())
selected_cases = random.sample(instance_names, 2)

print(f"Comparing instances: {selected_cases[0]} and {selected_cases[1]}")
diffs = diff_instance(data_in[selected_cases[0]], data_in[selected_cases[1]])
print("Differences between the two instances:")
for diff in diffs:
    print(diff)
```
